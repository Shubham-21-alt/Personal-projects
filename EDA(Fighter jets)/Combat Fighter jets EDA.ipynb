# Relationship Between Max speed,Max Altitude,Range

```python
# Calculate correlation between Max Speed and Range in Miles
correlation_speed_range = df['MaxSpeed'].corr(df['RangeinMiles'])

# Calculate correlation between Max Altitude and Range in Miles
correlation_altitude_range = df['MaxAltitude'].corr(df['RangeinMiles'])

# Calculate correlation between Max Speed and Max Altitude
correlation_speed_altitude = df['MaxSpeed'].corr(df['MaxAltitude'])

# Create a scatterplot grid
plt.figure(figsize=(18, 6))

# Scatterplot for Max Speed vs Range in Miles
plt.subplot(1, 3, 1)
sns.scatterplot(x='MaxSpeed', y='RangeinMiles', data=df)
plt.title(f'Max Speed vs Range in Miles\nCorrelation: {correlation_speed_range:.2f}')
plt.xlabel('Max Speed')
plt.ylabel('Range in Miles')

# Scatterplot for Max Altitude vs Range in Miles
plt.subplot(1, 3, 2)
sns.scatterplot(x='MaxAltitude', y='RangeinMiles', data=df)
plt.title(f'Max Altitude vs Range in Miles\nCorrelation: {correlation_altitude_range:.2f}')
plt.xlabel('Max Altitude')
plt.ylabel('Range in Miles')

# Scatterplot for Max Speed vs Max Altitude
plt.subplot(1, 3, 3)
sns.scatterplot(x='MaxSpeed', y='MaxAltitude', data=df)
plt.title(f'Max Speed vs Max Altitude\nCorrelation: {correlation_speed_altitude:.2f}')
plt.xlabel('Max Speed')
plt.ylabel('Max Altitude')

plt.tight_layout()
plt.show()

```

# Global Contribution to Jet Manufacturing

```python


# Countplot for Country of Origin
plt.figure(figsize=(20,8))
sns.countplot(x='CountryofOrigin', data=df, order=df['CountryofOrigin'].value_counts().index)
plt.title('Distribution of Combat Fighter Jets by Country of Origin')
plt.xlabel('Country of Origin')
plt.ylabel('Number of Jets Manufactured')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

# Impact of Generation on Performance Metrics

```python

# Create a boxplot layout for Max Speed, Range in Miles, and Max Altitude across Generations
plt.figure(figsize=(18, 6))

# Boxplot for Max Speed across Generations
plt.subplot(1, 3, 1)
sns.boxplot(x='Generation', y='MaxSpeed', data=df, order=df.groupby('Generation')['MaxSpeed'].median().sort_values(ascending=False).index)
plt.title('Max Speed Across Generations')
plt.xlabel('Generation')
plt.ylabel('Max Speed')
plt.xticks(rotation=45, ha='right')

# Boxplot for Range in Miles across Generations
plt.subplot(1, 3, 2)
sns.boxplot(x='Generation', y='RangeinMiles', data=df, order=df.groupby('Generation')['RangeinMiles'].median().sort_values(ascending=False).index)
plt.title('Range in Miles Across Generations')
plt.xlabel('Generation')
plt.ylabel('Range in Miles')
plt.xticks(rotation=45, ha='right')

# Boxplot for Max Altitude across Generations
plt.subplot(1, 3, 3)
sns.boxplot(x='Generation', y='MaxAltitude', data=df, order=df.groupby('Generation')['MaxAltitude'].median().sort_values(ascending=False).index)
plt.title('Max Altitude Across Generations')
plt.xlabel('Generation')
plt.ylabel('Max Altitude')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

# Trends in Engine Types

```python
# Count the occurrences of each Engine Type
engine_type_counts = df['EngineType'].value_counts()

# Create a pie chart for Engine Types with percentage values
plt.figure(figsize=(5, 5))
plt.pie(engine_type_counts, labels=engine_type_counts.index, autopct=lambda p: '{:.1f}%'.format(p) if p > 0 else '', startangle=90, colors=sns.color_palette('Set3'))
plt.title('Distribution of Engine Types in Combat Fighter Jets')
plt.show()
```

# Cost of jets vs Type of Engines

```python
plt.figure(figsize=(14, 8))
sns.violinplot(x='EngineType', y='Cost', data=df, order=df.groupby('EngineType')['Cost'].median().sort_values(ascending=False).index)
plt.title('Violin Plot of Price Across Engine Types')
plt.xlabel('Engine Type')
plt.ylabel('Cost')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

#  Correlation Heatmap 

```python
# Extracting numerical columns for correlation matrix
numerical_columns = ['CombatRange', 'Length(ft)', 'FuelCapacity', 'CrewSize', 'Cost',"Generation","Wingspan"]

# Calculate correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Plotting the heatmap
plt.figure(figsize=(6,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

# Stealth Technology vs Battles won

```python
plt.figure(figsize=(6, 6))
sns.pointplot(x='StealthTechnology', y='BattlesWon', data=df, errorbar=None)
plt.title('Stealth Technology vs Battles Won')
plt.xlabel('Stealth Technology')
plt.ylabel('Number of Battles Won')
plt.show()
```
